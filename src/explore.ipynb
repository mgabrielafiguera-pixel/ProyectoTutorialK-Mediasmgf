```python
# Your code here
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

df = pd.read_csv('https://breathecode.herokuapp.com/asset/internal-link?id=809&path=housing.csv')

df
```

```python

X = df[['Latitude', 'Longitude', 'MedInc']]

print(X.head())
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=6, random_state=42)

model.fit(X)

df['cluster'] = model.predict(X)

df
```

```python
print(df['cluster'].value_counts())
```

```python
model.cluster_centers_
```

```python
pca = PCA(n_components=3)

df_pca = pca.fit_transform(X)

df_pca = pd.DataFrame(df_pca,columns=['cp1','cp2','cp3'])

df_pca['cluster'] = df['cluster']
print(df_pca.head())
```

```python
centroides = pca.transform(model.cluster_centers_)
print("Coordenadas de los centroides en espacio PCA:")
print(centroides)
```

```python
import plotly.express as px

df_pca['cluster'] = df['cluster'].astype(str) 

print(df_pca.head())

fig = px.scatter_3d(df_pca, x='cp1', y='cp2', z='cp3',
              color='cluster')

fig.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

X = df.drop('cluster', axis=1)
y = df.cluster

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=42, train_size=0.8)

model_s = KNeighborsClassifier(n_neighbors=6)
model_s.fit(X_train, y_train)

y_pred = model_s.predict(X_test)
score = accuracy_score(y_test, y_pred)

print(f"Precisión del modelo corregido: {score}")
```

```python
import pickle
from pathlib import Path

Path("models").mkdir(parents=True, exist_ok=True)

with open("models/k-means_model_6.pkl", "wb") as f:
    pickle.dump(model, f)

with open("models/knn_classifier_6.pkl", "wb") as f:
    pickle.dump(model_s, f)

with open("models/scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

print("¡Modelos y escalador guardados exitosamente en la carpeta /models!")
```
