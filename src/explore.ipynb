```python
# Your code here

import pandas as pd

df = pd.read_csv('https://breathecode.herokuapp.com/asset/internal-link?id=809&path=housing.csv')

df
```

```python

X = df[['Latitude', 'Longitude', 'MedInc']]
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=6, random_state=42)

model.fit(df)

df['cluster'] = model.predict(df)

df
```

```python
model.cluster_centers_
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

pca = PCA(n_components=3)

df_pca = pca.fit_transform(df.drop('cluster',axis=1).values)

df_pca = pd.DataFrame(df_pca,columns=['cp1','cp2','cp3'])

df_pca
```

```python
centroides = pca.transform(model.cluster_centers_)
```

```python
X = df.drop('cluster',axis=1)
y = df.cluster

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,train_size=0.8)

from sklearn.neighbors import KNeighborsClassifier

model_s = KNeighborsClassifier(n_neighbors=6)

model_s.fit(X_train,y_train)

y_pred = model_s.predict(X_test)

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred )
```
